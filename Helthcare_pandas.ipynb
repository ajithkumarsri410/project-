```python
import pandas as pd
```

```python
df = pd.read_excel("/content/FINAL PROJECT DATA ANALYTICS Sheet 1.xlsx", sheet_name="Sheet1")
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['City'].fillna("Unknown", inplace=True)
df['Diagnosis_Code'].fillna("Not Provided", inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Gender_Code'] = df['Gender'].map({'Male': 0, 'Female': 1, 'Other': 2})
```

```python
def age_group(age):
    if age < 18:
        return "Child"
    elif 18 <= age <= 40:
        return "Adult"
    elif 41 <= age <= 60:
        return "Middle-Aged"
    else:
        return "Senior"

df['Age_Group'] = df['Age'].apply(age_group)
```

```python
def bill_category(amount):
    if amount > 10000:
        return "High"
    elif 5000 < amount <= 10000:
        return "Medium"
    else:
        return "Low"

df['Bill_Category'] = df['Bill_Amount'].apply(bill_category)
```

```python
total_visits_city = df['City'].value_counts()
print(total_visits_city)
```

```python
avg_bill_department = df.groupby('Department')['Bill_Amount'].mean()
print(avg_bill_department)
```

```python
insurance_counts = df['Count Insurance'].value_counts()
print(insurance_counts)
```
