```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/Cleaned_Healthcare_Data.xlsx")
```

```python
plt.figure(figsize=(10,6))
dept_visits = df['Department'].value_counts()
dept_visits.plot(kind='bar')
plt.title("Total Visits by Department")
plt.xlabel("Department")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,6))
df['Payment_Mode'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Distribution of Payment Modes")
plt.ylabel("")
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

bill_trend = df.groupby('Date')['Bill_Amount'].sum()

plt.figure(figsize=(10,6))
bill_trend.plot(kind='line', marker='o')
plt.title("Total Bill Amount Over Time")
plt.xlabel("Date")
plt.ylabel("Total Bill Amount ()")
plt.grid(True)
plt.tight_layout()
plt.show()
```
