```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel("/content/Healthcare Final.xlsx", sheet_name="Sheet1")
```

```python
df.columns
```

```python
Bill_amounts = df['Bill_Amount'].dropna()
print(Bill_amounts)
```

```python
Mean_bill = np.mean(Bill_amounts)
print(Mean_bill)
```

```python
Median_bill = np.median(Bill_amounts)
print(Median_bill)
```

```python
Mode_bill = Bill_amounts.mode()[0]
print(Mode_bill)
```

```python
Variance_bill = np.var(Bill_amounts)
print(Variance_bill)
```

```python
Std_bill = np.std(Bill_amounts)
print(Std_bill)
```

```python
Corr_age_bill = np.corrcoef(df['Age'], df['Bill_Amount'])[0,1]
print(Corr_age_bill)
```

```python
Q1 = np.percentile(Bill_amounts, 25)
Q3 = np.percentile(Bill_amounts, 75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
Outliers = df[(df['Bill_Amount'] < lower_limit) | (df['Bill_Amount'] > upper_limit)]
print(Outliers)
```

```python
lower_bound = Mean_bill - 2 * Std_bill
upper_bound = Mean_bill + 2 * Std_bill
Within_95 = df[(df['Bill_Amount'] >= lower_bound) & (df['Bill_Amount'] <= upper_bound)]
print(Within_95)
```
